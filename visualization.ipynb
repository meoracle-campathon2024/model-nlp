```python
Qwen15_500m = 0.64
bloom_560m = 0.48
Gemma_2B = 0.77
Qwen15_4B = 0.62
glm4_9B = 0.8

#plot bar chart to compare the accuracy of different models
import matplotlib.pyplot as plt

# data to plot
n_groups = 5
means = (Qwen15_500m, bloom_560m, Gemma_2B, Qwen15_4B, glm4_9B)

# create plot
fig, ax = plt.subplots()
index = range(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, means, bar_width,
alpha=opacity,
#color light blue
color='#7F27FF',
label='BLEU score')

plt.xlabel('Models')
plt.ylabel('BLEU score')
plt.title('BLEU score of different models')
plt.xticks(index, ('Qwen-1.5 0.5B', 'BLOOM 0.5B', 'Gemma 2B', 'Qwen-1.5 4B', 'GLM-4 9B'))
plt.legend()

plt.tight_layout()
plt.savefig('BLEU score of different models.png', dpi=600)
# plt.show()
```
